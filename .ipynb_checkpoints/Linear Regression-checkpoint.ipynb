```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
line = lambda x : (-2 / 3) * x + 14 / 3
data_range = np.array([-4, 10])
data_num = 100
noise_std = 0.5
```

```python
x = np.random.uniform(data_range[0], data_range[1], size = data_num)
y = line(x)
```

```python
xn = x + np.random.normal(scale = noise_std, size = x.shape)
yn = y + np.random.normal(scale = noise_std, size = y.shape)
```

```python
A = np.vstack((xn, np.ones(xn.shape))).T
b = yn
regression_line = np.linalg.pinv(A) @ b
```

```python
plt.title(f'Line : y = {regression_line[0] : .3f} * x + {regression_line[1] : .3f} ')
plt.plot(data_range, line(data_range), 'r-', label = 'Solid line')
plt.plot(xn, yn, 'b.', label = 'Noisy data')
plt.plot(data_range, regression_line[0] * data_range + regression_line[1], 'g-', label = 'Estimate')
#plt.xlim(data_range)
plt.legend()
plt.show()

plt.title(f'Line : y = {regression_line[0] : .3f} * x + {regression_line[1] : .3f} ')
plt.plot(data_range, line(data_range), 'r-', label = 'Solid line')
plt.plot(xn, yn, 'b.', label = 'Noisy data')
plt.plot(data_range, regression_line[0] * data_range + regression_line[1], 'g-', label = 'Estimate')
plt.xlim(data_range)
plt.legend()
plt.show()
```
