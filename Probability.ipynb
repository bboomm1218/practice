```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_range = np.array([-3, 3])
xs = np.linspace(*x_range, 1000)
mu = 0
sigma_set = [0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]

for sigma in sigma_set:
    pdf = 1 / sigma / np.sqrt(2 * np.pi) * np.exp(-0.5 * ((xs - mu) / sigma) ** 2)
    plt.plot(xs, pdf, label = f'N({mu : .1f}, {sigma : .2f})')

plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend(framealpha = 0.8)
plt.show()
```

```python
hist_pts = 10000
hist_bins = 100
dice_range = (1, 6)
n = 100
```

```python
samples = []
for i in range(hist_pts):
    samples.append(np.mean(np.random.randint(dice_range[0], dice_range[1] + 1, n)))

plt.title(f'Sample Mean (n = {n})')
plt.hist(samples, bins = hist_bins, range = dice_range, density = True, align = 'mid')
plt.xlim(dice_range[0], dice_range[1])
plt.yticks([])
plt.show()
```
