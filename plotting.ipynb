```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('data/class_score_en.csv', 'r') as f:
    reader = pd.read_csv(f)
    reader = np.array(reader)
    mid = []
    final = []
    total = []
    for m, f in reader:
        mid.append(m)
        final.append(f)
        total.append(m + f)
```

```python
with open('data/class_score_kr.csv', 'r') as f:
    reader = pd.read_csv(f)
    reader = np.array(reader)
    mid2 = []
    final2 = []
    total2 = []
    for m, f in reader:
        mid2.append(m)
        final2.append(f)
        total2.append(m + f)
```

```python
plt.scatter(mid, final, s = 30, c = 'blue', marker = '+')
plt.scatter(mid2, final2, s = 30, c = 'red', )
plt.xlabel('Midterm Scores')
plt.ylabel('Final Scores')
plt.axis('equal')
plt.grid()
plt.show()
```

```python
plt.hist(np.array(total) * (4 / 9), bins = 20, color = 'blue')
plt.hist(np.array(total2) * (4 / 9), bins = 20, color = 'red')
plt.xlabel('Total Scores')
plt.show()
```
